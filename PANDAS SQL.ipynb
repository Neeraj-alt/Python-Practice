```python
raw_data={
    'subject_id':['1','2','3','4'],
    'first_name':['Ata','Er','ERF','er'],
    'last_name':['er','rt','ty','uj']
    }
t=pd.DataFrame(raw_data,columns=['subject_id','first_name','last_name'])
t
```

```python
data2={
   'subject_id':['1','2','3','4'],
    'first_name':['errt','sdfgsg','asgag','asfasf'],
    'last_name':['dgs','fgdfgt','sdfy','sdfsj']
}
h=pd.DataFrame(data2,columns=['subject_id','first_name','last_name'])
h
```

```python
new=pd.concat([t,h])
new
```

```python
data3={
    'subject_id':['1','2','3','4','1','2','3','4'],
    'testid':['1','2','3','4','1','2','3','4']}
dr=pd.DataFrame(data3,columns=['subject_id','testid'])
dr
```

```python
q=pd.merge(new,dr)
q
```

```python
data_r={
    'Name':['qw','werty','tha','yah'],
    'Colour':['Pink','Orange','Pink','Blue'],
    'Brand':['Qwerty','Fog','Sputnik','ASD']}
E=pd.DataFrame(data_r,columns=['Name','Colour','Brand'])
E
```

```python
data_q={
    'Name':['qw','ergt','tha','hay'],
    'Colour':['Pink','Red','Pink','Blue'],
    'Brand':['Qwerty','Fog','Sputnik','ASD']}
d=pd.DataFrame(data_q,columns=['Name','Colour','Brand'])
d
```

```python
pd.merge(E,d,on='Colour',how='inner')
```

```python
pd.merge(E,d, on='Brand',how='outer')
```
