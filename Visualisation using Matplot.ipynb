```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
a=np.array([1,2,3,4,5,6])
b=np.array([1,2,3,4,5,6])
```

```python
plt.title('Linear PLot')
plt.xlabel('a')
plt.ylabel('b')
plt.plot(a,b)
plt.show()
```

```python
plt.plot(a,b,'--y')
plt.show()
```

```python
a=[1,2,3,4,5,6]
b=[12,3,45,56,23,12]
plt.bar(a,b)
plt.title('Bar Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
s=np.array([12,23,34,12,34,3,234,23,234,4,23,1323,4123,34])
plt.hist(s,bins=[10,100,200,300,400,500,600])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
